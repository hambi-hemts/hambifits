```python
def noisy(y, w = 1):
    for i, yval in enumerate(y):
        r = random.random()
        s = random.random()
        if s < 0.5:
            y[i] = yval + w*max(y)*r
        else:
            y[i] = yval - w*max(y)*r
    return y

def noisy(y, w = 1):
    for i, yval in enumerate(y):
        r = random.random()
        s = random.random()
        if s < 0.5:
            y[i] = yval + w*r
        else:
            y[i] = yval - w*r
    return y

def func(p, x):
    return p[0] * x
```
